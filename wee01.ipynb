If you wanted to find all the mentions of MY courses in this string,
then you could use any one of these regex patterns:
- MY[0-9][0-9][0-9]
- MY\d\d\d
- MY[0-9]+
- MY.{3}
- MY[^A-z]{3}

```python
import re
re.findall(r"MY\d+", "Dr Hübert teaches MY472 and MY459.")
```
